```python
import pandas as pd
import numpy as np
from pylab import *
import matplotlib.pyplot as plot


target_url = ("http://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data")
abalone = pd.read_csv(target_url,header=None,prefix="V")
abalone.columns = ['sex','Length','Diameter','Height','Whole weight','Shucked weight','Viscera weight','Shell weight','Rings']

#print(abalone)
print(abalone.head())
print(abalone.tail())

summary = abalone.describe()
print(summary)

# plot the box plot of the features
array = abalone.iloc[:,1:9].values
print(array)
boxplot(array)
plot.xlabel("feature index")
plot.ylabel("quartile ranges")
show()


# remove the last column and replot
array2 = abalone.iloc[:,1:8].values
boxplot(array2)
plot.xlabel("feature index")
plot.ylabel("quartile ranges")
show()

# we can also normalize the data
abalone_normalized = abalone.iloc[:,1:9]
#print(abalone_normalized)
for i in range(8):
    mean = summary.iloc[1,i]
    std = summary.iloc[2,i]
    abalone_normalized.iloc[:,i]=(abalone_normalized.iloc[:,i]-mean)/std
array3 = abalone_normalized.values
boxplot(array3)
plot.xlabel("feature index")
plot.ylabel("quartile ranges")
show()
```
